```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("C:\\Users\\talha\\Downloads\\ZOMATO\\zomato.csv")
```

```python
data.shape
```

```python
data.head()
```

```python
#Removing Uneccery Data
data = data.drop(['url','phone', 'address', 'listed_in(city)', 'menu_item','reviews_list','dish_liked'], axis=1)
```

```python
#Renaming columns Name
data = data.rename(columns={"name":"Restaurant Name","online_order":"Online Order","book_table":"Booking Type (Online)",
                       "rate":"Rating", "votes":"Votes","location":"Location","rest_type":"Restaurant Type",
                "cuisines":"Food Type", "approx_cost(for two people)":"Cost","listed_in(type)":"Type"})
data.head()
```

```python
data.head()
```

```python
data.info()
```

```python
data.duplicated().sum()
```

```python
data.describe()
```

```python
#Removing Duplicate Data
data= data.drop_duplicates()
```

```python
data.isna().sum()
```

```python
#Removing them due to their minimal presence
data_to_drop = ['Location', 'Restaurant Type', 'Food Type', 'Cost']
data = data.dropna(subset=data_to_drop)
```

```python
data['Rating'].unique()
```

```python
#Dealing with Unwanted values in Rating Column
data['Rating'] = pd.to_numeric(data['Rating'].replace({'NEW': np.nan, '/5': '' ,'-':np.nan}, regex=True))
```

```python
#Filling Blank values with Mean
data['Rating'] = data['Rating'].fillna(data['Rating'].mean())
```

```python
#Changing the data type to numeric from object data type
data['Cost'] = pd.to_numeric(data['Cost'], errors='coerce')

# Since Cost was for two Person, converting to one person
data['Cost'] = data['Cost']/2
```

```python
data.info()
```

```python
#No NaN value
data.isna().sum()
```

```python
data.shape
```

```python
plt.figure(figsize=(4,3))
sns.countplot(x='Online Order',data = data)
plt.title("No of Online Orders")
plt.show()
```

```python
plt.figure(figsize=(3,3))
plot = sns.countplot(x='Booking Type (Online)',data=data)
plt.title('No of User Order Online')
plt.show()
```

```python
plt.figure(figsize=(7, 5))

#filtering top 5 values
temp = data.groupby('Location')['Cost'].mean().sort_values(ascending=True).head(5)

sns.barplot(x=temp.index, y=temp)


plt.title('Average Cost for Top 5 Locations')
plt.xlabel('Location')
plt.ylabel('Mean Cost')

plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x= 'Type', data=data)
plt.title('Most Prefered Restaurant Type')
plt.show()
```

```python
#Filtering Data
df_plot = data.groupby('Booking Type (Online)')['Cost'].mean().sort_values(ascending = True)
plt.figure(figsize=(3, 4))
sns.barplot(x=df_plot.index,y=df_plot)
plt.title('Average Cost of Online and Offline Order')
plt.xlabel('Online Order')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.title("Best Location By Votes")
s=data.groupby("Location")[["Votes"]].sum().sort_values("Votes", ascending=False).head(20)
sns.barplot(s, x="Votes", y=s.index);
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title("Best Location By Ratings")

site = data.groupby("Location")[["Rating"]].sum().sort_values("Rating", ascending=False).head(20)
sns.barplot(site, y = site.index, x = "Rating");
```

```python
plt.figure(figsize=(10,3))
sns.countplot(data = data , x = 'Restaurant Type', order = data['Restaurant Type'].value_counts().iloc[:5].index)
plt.title('Most Preferable Restaurant Type')
plt.show()
```

```python
plt.figure(figsize=(20,5))
no_rest = data.groupby('Location')['Restaurant Type'].count().nlargest(10)

a = sns.barplot(x = no_rest.index, y = no_rest.values)

for i in a.containers:
    a.bar_label(i)

plt.title("Area Wise Restaurant")
plt.xlabel('locations')
plt.ylabel('numbers of restaurants')
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.kdeplot(data = data, x = 'Rating', shade=True)
plt.title('Ratings Distribution')
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(x='Type', y='Cost', data=data)
plt.show()
```
